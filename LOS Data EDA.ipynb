# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
# Read in the data

df = pd.read_csv('LOS Dataset.csv', header=0)
```

**Understanding the Data**

```python
# Review the first 30 rows of the dataset

pd.set_option('display.max_columns', None)
df.head(30)
```

```python
df.shape
```

```python
df.describe()
```

**Cleaning the Data**

```python
# Remove unnecessary columns 'DB Record', '% Purchase', '% Refi', '% Construction', '% MOD','% In Process', '% Approval', '% In Origination Review', '% In Closing' and 'Loan #'

df.columns
```

```python
df_subset = df[['Sales Team', 'Account', 'Loan Amt', 'Processor',
       'Application Date', 'Lock Expiration', 'Appraisal Date Ordered',
       'Appraisal Date Received', 'Estimated Fund Date',
       'Last Rate Lock Request', 'Apprisal Due Date', 'Month Fund', 'Product',
       'Loan Status', 'LOS System', 'Purpose', 'Days to Process',
       'Actual Process/Throughput', 'Days to Closing ', 'Processing Cost',
       'Closing Cost', 'Closer ', 'P&I', 'Type']]
df_subset
```

```python
# Check for missing values
 
df_subset.isnull().sum()
```

```python
# Overview of data types of each column in the dataframe

df_subset.dtypes
```

```python
# Reveiw all the number columns

df_subset.select_dtypes(include=['number'])
```

```python
# Review all the object columns

df_subset.select_dtypes(include=['object'])
```

```python
# Looking at the stats above, the negative values from 'Actual Process/Throughput' column needs to be removed.

df_subset.drop(df_subset[df_subset['Actual Process/Throughput']<0].index, inplace = True)
```

```python
df_subset.drop(df_subset[df_subset['Days to Closing ']<0].index, inplace = True)
```

```python
df_subset.describe()
```

```python
# Look for the correlation between variables

corr=df_subset.corr()
corr
```

```python
# Looking the heatmap below, 'Actual Process/Throughput' and 'Days to Process', and 'Processing Cost' and 'Days to Process' are highly correlated.

sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
```

```python
# Understand the distribution of the data

df_subset.hist(bins=60, figsize=(20,10))
```


**Let's group by the 'Actual Process/Throughput' and 'Loan Status' to get some insights on how many days it takes to process loans in each loan status.**
**Please note, the 'Actual Process/Throughput' was calculated by 'Estimated Fund Date' - 'Application Date'.**

```python
# Reveiw the stats of 'Actual Process/Throughput' by 'Loan Status' 

df_subset.groupby('Loan Status')['Actual Process/Throughput'].describe()
```

```python
# Create histograms using variable 'Loan Status' and 'Actual Process/Throughput' to undestand the distribution of those data
# Too little data for the 'Approved' and 'Review' to show in the histogram; however, 'In Closing' and 'In Processing' are both right skewed, and the majority of the loans are processed under 100 days
# Need to investigate outliers

g=sns.FacetGrid(df_subset, col='Loan Status')
g.map(plt.hist, 'Actual Process/Throughput')
```

```python
# Create a boxplot based on the loan status to discover outliers
# The majority of the loan closed within 100 days

sns.boxplot(x='Loan Status', y='Actual Process/Throughput', data=df_subset)
sns.stripplot(x='Loan Status', y='Actual Process/Throughput', data=df_subset,
             jitter=True,
             marker='o',
             alpha=0.8,
             color='black')
```

```python
# On average, mortgate loans close between 30 to 60 days. Let's look at loans that are processed within 60 days.

Under_60_days = df_subset[df_subset['Actual Process/Throughput']<60]
```

```python
Under_60_days.head(50)
```

```python
Under_60_days['Processing Cost'].max()
```

```python
# Loans that processed under 60 days, list out the transactions that had the highest origination fee?

Under_60_days[Under_60_days['Processing Cost'] == Under_60_days['Processing Cost'].max()]
```

```python
# Loans that processed under 60 days, list out teams that processed the most transactions that had the highest origination fee?

MaxFee = Under_60_days[Under_60_days['Processing Cost'] == Under_60_days['Processing Cost'].max()]
MaxFee['Sales Team'].value_counts()
```
